```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#Loading dataset
df = pd.read_csv('/content/RawStudentperformance.csv')
```

```python
df
```

```python
#Checking null values
null_values = df.isnull().sum()

# Display the number of null values per column
print("Null values in each column:")
print(null_values)
```

#Bar Chart

```python
plt.figure(figsize=(10,6))
avg_scores_by_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
avg_scores_by_gender.plot(kind='bar')
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

#Pie Chart

```python
plt.figure(figsize=(8,8))
parental_education = df['parental level of education'].value_counts()
plt.pie(parental_education, labels=parental_education.index, autopct='%1.1f%%', startangle=140)
plt.title('Parental Level of Education Distribution')
plt.show()
```

#Box Chart

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['math score', 'reading score', 'writing score']])
plt.title('Score Distribution for Math, Reading, and Writing')
plt.ylabel('Score')
plt.show()
```

#Heat Map

```python
plt.figure(figsize=(10,6))
corr = df[['math score', 'reading score', 'writing score']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Between Math, Reading, and Writing Scores')
plt.show()
```

#Scatter Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(df['math score'], df['writing score'], label='Math vs Writing Score', alpha=0.5, color='green')
plt.title('Math Score vs Writing Score')
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.legend()
plt.show()
```

#Bar Chart - Average Scores by Gender: This bar chart compares the average performance of male and female students in math, reading, and writing. It helps to highlight any gender-based differences or similarities in subject performance.

#Pie Chart - Parental Level of Education: This pie chart shows the distribution of students based on their parents' educational attainment. It provides an overview of the variety of educational backgrounds among the students' families.

#Box Plot - Score Distribution for Each Subject: The box plot displays the distribution of scores for each subject, showing the median, quartiles, and potential outliers. It provides a clear picture of the spread and variation in math, reading, and writing scores.

#Heatmap - Correlation Between Variables: The heatmap reveals the strength and direction of correlations between math, reading, and writing scores. It allows for a quick understanding of how performance in one subject might relate to performance in others.

#Scatter Plot - Math Score vs. Writing Score: The scatter plot visualizes the relationship between math and writing scores across individual students. It helps to uncover any patterns or correlations in performance between these two subjects.
