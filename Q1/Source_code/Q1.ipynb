```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading dataset
data = pd.read_csv('/content/Frailtyraw.csv')
```

```python
#Display data
data
```

```python
#Showing if there is any missing values in the data
missing_values = data.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
#Checking null values
null_values = data.isnull().sum()

# Display the number of null values per column
print("Null values in the column:")
print(null_values)
```

```python
#Graph
plt.figure(figsize=(10, 6))
plt.scatter(data['Height'], data['Weight'], color='blue')
plt.title('Height vs Weight')
plt.xlabel('Height (inches)')
plt.ylabel('Weight (lbs)')
plt.show()
```

The scatter plot visually shows the relationship between height and weight. Although a general trend is visible, individual differences and potential outliers indicate that this relationship is complex and affected by various factors.

```python
# Pie Chart
plt.figure(figsize=(8, 8))
frailty_counts = data['Frailty'].value_counts()
plt.pie(frailty_counts, labels=frailty_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Frailty')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

```

Prevalence of Frailty: 40% of the individuals in the sample are classified as frail (Y).

Prevalence of Non-Frailty: 60% of the individuals are not frail (N).
